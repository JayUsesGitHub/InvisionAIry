```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.image import imread
import pydot
import cv2
import warnings
import json

# Suppress warnings
warnings.filterwarnings('ignore')

# TensorFlow and Keras libraries for image processing and model training
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import plot_model
from tensorflow.keras import backend
from sklearn.metrics import confusion_matrix, classification_report

# SHAP for model explainability
import shap
from operator import itemgetter

# Holoviews for interactive plots (optional, depending on your needs)
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
```

```python
data_dir = r'./InvisionAIry_Personal_Projects/Mechanical_Quality_Assurance'

# Using os.path.join to handle file paths correctly across platforms
my_data_dir = os.path.join(data_dir, r'real-life-industrial-dataset-of-casting-product', r'casting_data', r'casting_data')
train_path = os.path.join(my_data_dir, r'train')
test_path = os.path.join(my_data_dir, r'test')
```

```python
# Define image preprocessing
image_gen = ImageDataGenerator(
    rescale=1/255,  # Normalize pixel values to the range [0, 1]
    zoom_range=0.1,  # Random zoom
    brightness_range=[0.9, 1.0]  # Random brightness adjustment
)
```

```python
# Set image shape and batch size
image_shape = (300, 300, 1)  # Image dimensions (300x300 grayscale images)
batch_size = 32
```

```python
# Ensure directories exist before using them
if not os.path.exists(train_path) or not os.path.exists(test_path):
    print("Warning: One or both of the directories don't exist. Please check the paths.")
else:
    # Create training data generator
    train_set = image_gen.flow_from_directory(
        train_path,
        target_size=image_shape[:2],  # Resize images to match the target size
        color_mode="grayscale",  # Ensure the images are loaded in grayscale
        classes=['def_front', 'ok_front'],  # Class names should be in the correct order
        batch_size=batch_size,  # Batch size
        class_mode='binary',  # Binary classification: 'def_front' vs. 'ok_front'
        shuffle=True,  # Shuffle images randomly
        seed=0  # Random seed for reproducibility
    )

    # Create test data generator
    test_set = image_gen.flow_from_directory(
        test_path,
        target_size=image_shape[:2],  # Resize images to match the target size
        color_mode="grayscale",  # Ensure the images are loaded in grayscale
        classes=['def_front', 'ok_front'],  # Class names should be in the correct order
        batch_size=batch_size,  # Batch size
        class_mode='binary',  # Binary classification
        shuffle=False,  # Don't shuffle test data
        seed=0  # Random seed for reproducibility
    )

    # Check the class indices to verify the mapping
    print("Class indices:", train_set.class_indices)
```

```python
# Clear previous sessions to avoid cluttering the model
backend.clear_session()
```

```python
# Initialize the model
model = Sequential()

# First Conv2D layer with input shape defined
model.add(Conv2D(filters=16, kernel_size=(7,7), strides=2, input_shape=image_shape, activation='relu', padding='same'))

# First MaxPooling layer
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))

# Second Conv2D layer, no need to redefine input_shape
model.add(Conv2D(filters=32, kernel_size=(3,3), strides=1, activation='relu', padding='same'))

# Second MaxPooling layer
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))

# Third Conv2D layer, no need to redefine input_shape
model.add(Conv2D(filters=64, kernel_size=(3,3), strides=1, activation='relu', padding='same'))

# Third MaxPooling layer
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))

# Flatten the output to feed into the Dense layer
model.add(Flatten())

# Fully connected Dense layer
model.add(Dense(units=224, activation='relu'))

# Dropout layer for regularization
model.add(Dropout(rate=0.2))

# Output layer with a sigmoid activation for binary classification
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Model summary
model.summary()
```

```python
# Visualizing the model
plot_model(model, show_shapes=True, expand_nested=True, dpi=60)
```
